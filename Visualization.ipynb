```python
bb = data.DataReader(name='BB',data_source='yahoo',start='2007-1-1',end='2020-12-31')
bb.head(3)
```

```python
def rank(stock_price):
    if stock_price <=10:
        return 'Poor'
    elif stock_price <=50:
        return 'Satisfactory'
    else:
        return 'Good'
```

```python
bb['Close'].apply(rank).value_counts().plot(kind='bar')
```

```python
bb = data.DataReader(name='BB',data_source='yahoo',start='2007-1-1',end='2020-12-31')
bb.head(3)
```

```python
bb['Close'].mean()
```

```python
def rank(stock_price):
    if stock_price >=29.452764073094578:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
bb['Close'].apply(rank).value_counts().plot(kind='pie',legend=True)
```
